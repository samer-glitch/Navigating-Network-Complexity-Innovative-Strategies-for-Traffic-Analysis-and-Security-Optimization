```python
# In machine learning, cardinality refers to the uniqueness of data values within a column.
# Specifically, it is the count of distinct values in a feature.
# High-cardinality attributes have a large number of unique values,
# while low-cardinality attributes have a small number of unique valuesCardinality is an important factor to consider during feature selection and preprocessing,
# as high-cardinality features can significantly increase the complexity of a model, potentially leading to overfitting or increased computational demand.

import pandas as pd

# List of categorical columns to check for cardinality
categorical_columns = ['proto', 'src_ip', 'dst_ip', 'category', 'application_protocol', 'web_service']

# Check cardinality for each categorical column
for column in categorical_columns:
    cardinality = df2[column].nunique()
    print(f"{column}: Cardinality = {cardinality}")
```
